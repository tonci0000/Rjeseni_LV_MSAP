```python
import kagglehub
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import seaborn as sn
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

1. Učitavanje i pregled podataka

```python
data = pd.read_csv("AirQualityUCI.csv")
print(data.info())
print(data.head(5))

data = data[["Date", "Time", "CO(GT)", "PT08.S1(CO)", "NMHC(GT)", "C6H6(GT)", "PT08.S2(NMHC)", "NO2(GT)", "PT08.S4(NO2)", "PT08.S5(O3)", "T", "RH", "AH"]]
```

```python
print(data.describe())
data.dtypes
```

U datasetu se nalaze vrijednosti tipa float (osim varijabli Date i Time). Npr. postoji 9357 NO2 varijabli čija je aritmetička sredina -34.21 uz standardnu devijaciju 126.94. Najmanja izmjerena razina NO2 je -200. a najveća 340. Uz to imamo i pregled kvartila.

```python
data.isna().sum()
```

Ovdje vidimo nedostajuće vrijednosti.

```python
data = data.replace(-200, np.nan)
print(data.isna().sum())
data1 = data.copy()
```

Zamjena vrijednosti -200 s NaN dovela je do puno većeg broja nedostajućih vrijednosti u varijablama CO, NMHC i NO2.

2. Definiranje hipoteza i odabir varijabli za regresiju

```python
# Odabir varijabli CO(GT) (ugljikov monoksid, zavisna varijabla y) i C6H6(GT) (benzen, nezavisna varijabla x) - nedostajuće vrijednosti su uklonjene

data_cleaned1 = data.dropna(subset=["CO(GT)", "C6H6(GT)"])
print(data_cleaned1.isna().sum())
data_cleaned1 = data_cleaned1[["CO(GT)", "C6H6(GT)"]]
print(data_cleaned1.head())

# Odabir varijabli za drugi model NO2(GT) (koncentracija dušikovog dioksida, zavisna varijabla y) i C6H6(GT) (benzen, nezavisna varijabla x) - nedostajuće vrijednosti su uklonjene

data_cleaned2 = data1.dropna(subset=['NO2(GT)', 'C6H6(GT)'])
print(data_cleaned2.isna().sum())
data_cleaned2 = data_cleaned2[['NO2(GT)', 'C6H6(GT)']]
print(data_cleaned2.head())
```

3. Izgradnja linearnog regresijskog modela

```python
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(data_cleaned1["C6H6(GT)"], data_cleaned1["CO(GT)"])
intercept_err1 = stats.linregress(data_cleaned1["C6H6(GT)"], data_cleaned1["CO(GT)"]).intercept_stderr
print(p_value1)

if p_value1 < 0.05:
    print("Postoji značajna linearna veza između koncentracije ugljikovog monoksida i benzena (odbacujemo nultu hipotezu).")
else:
    print("Nema dovoljno dokaza za postojanje značajne linearne veze koncentracije ugljikovog monoksida i benzena (ne odbacujemo nultu hipotezu).")

# Dodavanje regresijskog pravca

y_pred1 = intercept1 + slope1 * data_cleaned1["C6H6(GT)"]

plt.scatter(data_cleaned1["C6H6(GT)"], data_cleaned1["CO(GT)"], color='blue', label='Podaci', alpha=0.4)
plt.plot(data_cleaned1["C6H6(GT)"], y_pred1, color='red', label='Regresijski pravac')
plt.xlabel('Koncentracija benzena')
plt.ylabel('Koncentracija ugljikovog monoksida')
plt.title('Model A')
plt.legend()
plt.grid()
plt.show()
```

```python
print("Model A\n")
print(f"Nagib (slope): {slope1}")
print(f"Značenje nagiba: Koncentracija ugljikovog monoksida prosječno se mijenja za {slope1:.4f} za svako povećanje koncentracije benzena za 1 jedinicu.")
print(f"Odsječak (intercept): Kada je koncentracije benzena 0, onda je ugljikovog monoksida {intercept1}")
print(f"R^2: {r_value1**2:.4f}")
print("R^2 govori koliki postotak varijance u Y se može objasniti X-om, u ovom slučaju puno.")
print(f"Jednadžba: yA = {intercept1:.4f} + {slope1:.4f}*x1")
```

```python
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(data_cleaned2["NO2(GT)"], data_cleaned2['C6H6(GT)'])
intercept_err2 = stats.linregress(data_cleaned2["NO2(GT)"], data_cleaned2['C6H6(GT)']).intercept_stderr

print(p_value2)
if p_value2 < 0.05:
    print("Postoji značajna linearna veza između koncentracije dušikovog dioksida i benzena (odbacujemo nultu hipotezu).")
else:
    print("Nema dovoljno dokaza za postojanje značajne linearne veze koncentracije dušikovog dioksida i benzena (ne odbacujemo nultu hipotezu).")


y_pred2 = intercept2 + slope2 * data_cleaned2["NO2(GT)"]

plt.scatter(data_cleaned2["NO2(GT)"], data_cleaned2['C6H6(GT)'], color='blue', label='Podaci', alpha=0.4)
plt.plot(data_cleaned2["NO2(GT)"], y_pred2, color='red', label='Regresijski pravac')
plt.xlabel('Koncentracija dušikovog dioksida')
plt.ylabel('Koncentracija benzena')
plt.title('Model B')
plt.legend()
plt.grid()
plt.show()
```

```python
print("Model B\n")
print(f"Nagib (slope): {slope2:.4f}")
print(f"Značenje nagiba: Koncentracija benzena prosječno se mijenja za {slope2:.4f} za svako povećanje koncentracije dušikovog dioksida za 1 jedinicu.")
print(f"Odsječak (intercept): Kada je koncentracije dušikovog dioksida 0, onda je koncentracija benzena {intercept2:.4f}")
print(f"R^2: {r_value2**2:.4f}")
print("R^2 govori koliki postotak varijance u Y se može objasniti X-om, u ovom slučaju malo.")
print(f"Jednadžba: yB = {intercept2:.4f} + {slope2:.4f}*x2")
```

4. Intervali povjerenja za parametre

MODEL A


```python
confidence_level = 0.95
n = len(data_cleaned1)
t_critical1 = stats.t.ppf((1 + confidence_level) / 2, df=n-2)
margin_of_error1 = t_critical1 * std_err1
lower_bound1 = slope1 - margin_of_error1
upper_bound1 = slope1 + margin_of_error1

# Interpretacija intervala povjerenja
print("S 95% sigurnošću vjerujemo da je stvarni nagib između {:.4f} i {:.4f}.".format(lower_bound1, upper_bound1))

n = len(data_cleaned1)
t_critical11 = stats.t.ppf((1 + confidence_level) / 2, df=n-2)
margin_of_error11 = t_critical11 * intercept_err1
lower_bound11 = intercept1 - margin_of_error11
upper_bound11 = intercept1 + margin_of_error11
print(f"S 95% sigurnosti možemo tvrditi da se intercept (odsječak na osi y) modela A nalazi u intervalu: <{lower_bound11:.4f}, {upper_bound11:.4f}>")

# Interval ne uključuje 0 - nagib je značajan, postoji linearna veza između varijabli.
```

MODEL B

```python
confidence_level = 0.95
n = len(data_cleaned2)
t_critical2 = stats.t.ppf((1 + confidence_level) / 2, df=n-2)
margin_of_error2 = t_critical2 * std_err2
lower_bound2 = slope2 - margin_of_error2
upper_bound2 = slope2 + margin_of_error2

# Interpretacija intervala povjerenja
print("S 95% sigurnošću vjerujemo da je stvarni nagib između {:.4f} i {:.4f}.".format(lower_bound2, upper_bound2))

n = len(data_cleaned2)
t_critical22 = stats.t.ppf((1 + confidence_level) / 2, df=n-2)
margin_of_error22 = t_critical22 * intercept_err2
lower_bound22 = intercept2 - margin_of_error22
upper_bound22 = intercept2 + margin_of_error22
print(f"S 95% sigurnosti možemo tvrditi da se intercept (odsječak na osi y) modela B nalazi u intervalu: <{lower_bound22:.4f}, {upper_bound22:.4f}>")

# Interval uključuje 0 - nagib nije značajan, ne postoji značajna linearna veza između varijabli.
```

ako interval povjerenja za b1 sadrži nulu (0), tada:

ne odbacujemo nul-hipotezu, kažemo da nagib nije statistički značajan na toj razini pouzdanosti jer nema dokaza o linearnoj vezi između x i y (u populaciji)

Drugim riječima: podaci nisu dovoljno jaki da pokažu da se y sustavno mijenja s x.

Ovo je ekvivalentno t-testu za koeficijent: ako je p-vrijednost > α (npr. 0.05)
tada 95% CI sadrži nulu i zaključak je isti: koeficijent nije značajan

5. Analiza adekvatnosti modela

MODEL B


```python
# Reziduali
# data_cleaned2["C6H6(GT)"], data_cleaned2['NO2(GT)']
residuals2 = data_cleaned2['C6H6(GT)'] - (intercept2 + slope2 * data_cleaned2['NO2(GT)'])
fitted_values2 = intercept2 + slope2 * data_cleaned2['NO2(GT)']

plt.scatter(fitted_values2, residuals2, alpha=0.4)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predviđene vrijednosti')
plt.ylabel('Reziduali')
plt.title('Reziduali vs Predviđene vrijednosti')
plt.grid()
plt.show()
print("Reziduali nemaju sličnu raspršenost duž cijelog raspona predviđenih vrijednosti. Prema tome, pretpostavlja se da homoskedastičnost nije zadovoljena.")

sns.histplot(residuals2, kde=True)
plt.xlabel('Reziduali')
plt.title('Histogram reziduala')
plt.show()
print("Vrijednosti reziduala je blizu normalne distribucije.")


plt.figure()
stats.probplot(residuals2, dist="norm", plot=plt)
plt.title('Normalni Q-Q graf')
plt.show()
print("Većina reziduala pada blizu dijagonale na Q–Q grafu, što znači da su približno normalno raspoređeni. Repovi pomalo odstupaju.")
```

6. Koeficijent korelacije

MODEL B


```python
# data_cleaned2["C6H6(GT)"], data_cleaned2['NO2(GT)']

# Testiranje hipoteze H0: ρ = 0 pomoću t-statistike

r_value, p_value_scipy = stats.pearsonr(data_cleaned2["NO2(GT)"], data_cleaned2['C6H6(GT)'])

# Pearsonov koeficijent korelacije
print(f"Pearsonov koeficijent korelacije r = {r_value:.4f}")
print(f"p-vrijednost (iz scipy.pearsonr) = {p_value_scipy:.6f}")
print(f"Koeficijent determinacije (R^2) = {(r_value*r_value):.4f}")

# Testiranje hipoteze H0: ρ = 0 pomoću t-statistike
t_stat = r_value * np.sqrt((n - 2) / (1 - r_value**2))
df = n - 2
p_value_t = 2 * (1 - stats.t.cdf(abs(t_stat), df=df))

print(f"\nt-statistika = {t_stat:.4f}")
print(f"Stupnjevi slobode = {df}")
print(f"p-vrijednost (iz t-statistike) = {p_value_t:.6f}")

# Odluka na razini značajnosti 5%
if p_value_t < 0.05:
    print("\nZaključak: p < 0.05 → odbacujemo H0.")
    print("Postoji statistički značajna linearna veza između benzena i dušikovog dioksida.")
else:
    print("\nZaključak: p ≥ 0.05 → ne odbacujemo H0.")
    print("Nema dovoljno dokaza za značajnu linearnu vezu između benzena i dušikovog dioksida.")

# Navedene varijable su umjereno i pozitivno povezane (R^2 = 0.37) (dok raste vrijednost dušikovog dioksida, raste i vrijednost benzena)
```

7. Testiranje adekvatnosti modela – F-test

MODEL B

```python
X = sm.add_constant(data_cleaned2["NO2(GT)"])  # dodaj presjek
model = sm.OLS(data_cleaned2['C6H6(GT)'], X).fit()

# Ispis cijelog modela
print(model.summary())

# Izdvajanje F-test informacija
F_statistic = model.fvalue
p_value_F = model.f_pvalue

print(f"\nF-statistika: {F_statistic:.4f}")
print(f"p-vrijednost (F-test): {p_value_F:.6f}")

# Odluka na razini značajnosti 5%
alpha = 0.05
if p_value_F < alpha:
    print("Zaključak: Model je statistički značajan (odbacujemo H0).")
else:
    print("Zaključak: Model nije statistički značajan (ne odbacujemo H0).")
```

U navedenim modelima, prisutna je jača pozitivna korelacija između varijabli u modelu A - kada raste vrijednost DUŠIKOVOG DIOKSIDA, raste i vrijednost benzena. Točnije, za svako povećanje razine DUŠIKOVOG DIOKSIDA za 1, vrijednost benzena se povećava za odrđen iznos. Oba modela su adekvatna prema F statističkom testu zbog čije smo p-vrijednosti imali dovoljno dokaza da odbacimo hipotezu da modeli nisu adekvatni.
