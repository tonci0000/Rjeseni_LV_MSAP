```python
#WELCHOVA ANOVA

import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.oneway import anova_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd

# Grupiranje podataka
groups = [
    group["math score"].values
    for name, group in df.groupby("parental level of education")
]

# Shapiro–Wilk test normalnosti
print("\nShapiro–Wilk test normalnosti:")
for name, group in df.groupby("parental level of education"):
    stat, p = stats.shapiro(group["math score"])
    print(f"{name}: W={stat:.3f}, p={p:.4f}")

# Leveneov test (pokazuje da su varijance različite)
levene_stat, levene_p = stats.levene(*groups)
print("\nLeveneov test homogenosti varijanci:")
print(f"Statistika={levene_stat:.3f}, p={levene_p:.4f}")

# -------------------------
# WELCHOVA ANOVA
# -------------------------
welch = anova_oneway(
    groups,
    use_var="unequal"   # KLJUČNO → Welchova ANOVA
)

print("\nWelchova ANOVA:")
print(f"F = {welch.statistic:.3f}")
print(f"p = {welch.pvalue:.5f}")

#Games–Howell post-hoc
import pingouin as pg

games_howell = pg.pairwise_gameshowell(
    data=df,
    dv="math score",
    between="parental level of education"
)

print("\nGames–Howell post-hoc test:")
print(games_howell)
```

```python
#RM Two-way ANOVA

#RM Two-Way ANOVA se koristi kada:

#1 zavisna varijabla (DV) – kontinuirana

#2 faktora

#barem 1 within-subjects (ponovljena mjerenja)

#drugi može biti:

#within → two-way RM ANOVA

#between → mixed ANOVA

#zavisni uzorci (unutar within faktora)

#normalnost po ćelijama

#sferičnost (za within faktore s ≥3 razine)

subject | time     | method | math_score
----------------------------------------
1       | before   | A      | 65
1       | after    | A      | 72
1       | followup | A      | 70
2       | before   | B      | 60

import scipy.stats as stats

print("\nShapiro–Wilk test normalnosti:")
for name, group in df.groupby(["time", "method"]):
    stat, p = stats.shapiro(group["math_score"])
    print(f"{name}: W={stat:.3f}, p={p:.4f}")

import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM

rm_anova = AnovaRM(
    data=df,
    depvar="math_score",
    subject="subject",
    within=["time"],
    between=["method"]   # ako ima between faktor
)

results = rm_anova.fit()

print("\nRM Two-Way ANOVA (Mixed design):")
print(results)

import pingouin as pg

spher = pg.sphericity(
    data=df,
    dv="math_score",
    subject="subject",
    within="time"
)

print("\nMauchlyjev test sferičnosti:")
print(spher)

posthoc_within = pg.pairwise_ttests(
    data=df,
    dv="math_score",
    within="time",
    subject="subject",
    padjust="bonferroni"
)

print("\nPost-hoc within faktor (Bonferroni):")
print(posthoc_within)

posthoc_between = pg.pairwise_ttests(
    data=df,
    dv="math_score",
    between="method",
    padjust="bonferroni"
)

print("\nPost-hoc between faktor:")
print(posthoc_between)

# primjer: usporedba metoda unutar svakog vremena
simple_effects = pg.pairwise_ttests(
    data=df,
    dv="math_score",
    between="method",
    within="time",
    subject="subject",
    padjust="bonferroni"
)

print("\nSimple effects (interakcija):")
print(simple_effects)
```
